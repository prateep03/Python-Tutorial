```python
# Clustering
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
# split the train data into independent variables (X) and dependent variable (Y)
x_iris = iris.drop('species', axis = 1)
print(x_iris.shape)
y_iris = iris['species']
print(y_iris.shape)
print(type(x_iris))
print(type(y_iris))
```

```python
x_iris.head()
```

```python
# apply KMeans model to group the data
from sklearn.cluster import KMeans
KMC = KMeans(n_clusters = 3)
KMC.fit(x_iris)
y_KMC = KMC.predict(x_iris)
```

```python
print(KMC.cluster_centers_)
```

```python
print(KMC.inertia_)
```

```python
# include new column 'cluster' to iris data
iris ["cluster"] = y_KMC
print(iris.head())
```

```python
print(type(iris))
```

```python
import numpy as np
withinss1 = 0
ratio = list()
for i in np.arange(1,11):
    KMC = KMeans(n_clusters = i)
    KMC.fit(x_iris)
    if i == 1:
        withinss1 = KMC.inertia_
    ratio.append (KMC.inertia_ / withinss1)
print(ratio)
```

```python
import matplotlib.pyplot as plt
plt.scatter(np.arange(1,11),ratio)
```
